#Stage1: Processing the dataset.

```python
#taking sub of dataset
df = df.sample(n=500, random_state=42)
df.shape
```

```python
# checking(NaN values).
print (f"missing_values count: ")
missing_values = df.isnull().sum()
missing_values
```

```python
# Removing rows with NaN in these columns
df.dropna(subset=['name','genre', 'location'], inplace=True)

# Filling missing values
df['best_comment'] = df['best_comment'].fillna(" ")
df['review_count'] = df['review_count'].fillna(0)

# Fill missing Names
df['name'] = df['name'].fillna('Unknown')
```

```python
# Rechecking(NaN values).
print (f"missing_values count: ")
missing_values = df.isnull().sum()
missing_values
```

```python
#1.Remove an unneeded column
df= df.drop(columns=['Unnamed: 0'])
print(df.columns)
```

```python
#Removing duplicate records.

print(f"duplicate records: {df.duplicated().sum()}")
df.describe()

print(f"dublicate count before removing: {df.duplicated().sum()}")

df = df.drop_duplicates(subset=['name', 'genre', 'location'])

print(f"dublicate count after removing: {df.duplicated().sum()}")
df
```

```python
 #Tokenizing text data:
df.loc[:, 'tokenized_comment'] = df['best_comment'].astype(str).apply(lambda x: x.split())
```

```python
df[['best_comment', 'tokenized_comment']].head()
```

```python
# Convert rating and review_count to numeric
df.loc[:, 'rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.loc[:, 'review_count'] = pd.to_numeric(df['review_count'], errors='coerce')

# Fill NaN if needed
df.loc[:, 'rating'] = df['rating'].fillna(df['rating'].mean())
df.loc[:, 'review_count'] = df['review_count'].fillna(0)

# Normalizing rating
min_rating = df['rating'].min()
max_rating = df['rating'].max()
df.loc[:, 'rating_scaled'] = (df['rating'] - min_rating) / (max_rating - min_rating)

# Normalizing review_count
min_review = df['review_count'].min()
max_review = df['review_count'].max()
df.loc[:, 'review_scaled'] = (df['review_count'] - min_review) / (max_review - min_review)

# Show results
df[['rating', 'rating_scaled', 'review_count', 'review_scaled']].head()
```

```python
#Validating data:
#Validate rating values (must be between 0 and 5)
rating_errors = df[(df['rating'] < 0)|(df['rating'] > 5)]
print(f'Rating issues (outside 0–5) {rating_errors}')

#Validate review_count (must be >= 0)
review_count_errors = df[df['review_count'] < 0]
print(f'Review count issues (negative values): {review_count_errors}')
```

```python
df[['rating', 'rating_scaled', 'review_count', 'review_scaled']].head()
```

```python
# Reformatting data:

# 1. Renaming columns
df = df.rename(columns={
    'name': 'Name',
    'rating': 'Rating',
    'review_count': 'Review_count',
    'genre': 'Genre',
    'location': 'Location',
    'best_comment': 'Best_comment'
})




# 1. Reformatting Review_count
def convert_review_count(review):

    review = str(review).strip().upper()
    if 'K' in review:

        return float(review.replace('K','')) * 1000
    elif 'M' in review:

        return float(review.replace('M','')) * 1000000

    elif review.replace('.', '', 1).isdigit():
        return float(review)
    else:
        return np.nan


df.loc[:, 'Review_count'] = df['Review_count'].apply(convert_review_count)


df.loc[:, 'Review_count'] = df['Review_count'].fillna(0).astype(int)
print("1. Done Reformatting 'Review_count'")


# 1. Reformatting Rating
df.loc[:, 'Rating'] = pd.to_numeric(df['Rating'], errors='coerce')


mean_rating = df['Rating'].mean()
df.loc[:, 'Rating'] = df['Rating'].fillna(mean_rating)
print(f"2. Done Reformatting 'Rating' {mean_rating:.2f}.")



# 3. convert Review_count to numeric
df['Review_count'] = pd.to_numeric(df['Review_count'], errors='coerce').fillna(0).astype(int)



# 4. convert Rating to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
mean_rating = df['Rating'].mean()
df['Rating'] = df['Rating'].fillna(mean_rating)
df['Rating'] = df['Rating'].astype(float)


# 5. (Renormalizing/ReScaling)
min_rating = df['Rating'].min()
max_rating = df['Rating'].max()

df['rating_scaled'] = (df['Rating'] - min_rating) / (max_rating - min_rating)



print("\nChecking values are updated:")
df.info()
```

# Stage2: Data Analysis.

```python
# Finding central values

# Mean
mean_rating = df["Rating"].mean()
print("Mean Rating:", mean_rating)

# Median
median_rating = df["Rating"].median()
print("Median Rating:", median_rating)

# Mode
mode_rating = df["Rating"].mode()
print("Mode Rating:", mode_rating)


#Finding measures of dispersion

# Standard Deviation
std_rating = df["Rating"].std()
print("Standard Deviation:", std_rating)

# Variance
variance_rating = df["Rating"].var()
print("Variance:", variance_rating)

# Number of Comments
reviews_count = df["Best_comment"].count()
print("Number of Comments:", reviews_count)
```

```python
#Finding common words in text data
all_comments = " ".join(df["best_comment"].astype(str))
words = all_comments.split()

word_counts = {}

for w in words:
    w = w.strip(".,!?\"()[]{};:-")
    if w != "" and w.lower() != "nan":
        word_counts[w] = word_counts.get(w, 0) + 1

word_counts_df = pd.DataFrame(word_counts.items(), columns=["word", "count"])
word_counts_df = word_counts_df.sort_values("count", ascending=False)

word_counts_df.head(10)
```

```python
#Grouping and aggregating data first part :
mean_rating_map = df.groupby('Name')['Rating'].mean()

df['Mean_Rating_Per_Event'] = df['Name'].map(mean_rating_map)


global_mean_rating = df['Rating'].mean()
df['Mean_Rating_Per_Event'] = df['Mean_Rating_Per_Event'].fillna(global_mean_rating)
```

```python
#Calculating the correlation between variables.
df['Mean_Rating_Per_Event'] = pd.to_numeric(df['Mean_Rating_Per_Event'], errors='coerce')


correlation_matrix = df[['Rating', 'Mean_Rating_Per_Event']].corr()


print(correlation_matrix)


rating_mean_correlation = correlation_matrix.loc['Rating', 'Mean_Rating_Per_Event']
print(f"\n corelation value  {rating_mean_correlation:.4f}")
```

```python
numeric_cols = ['Rating', 'Review_count', 'rating_scaled', 'Mean_Rating_Per_Event']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

correlation_matrix_full = df[numeric_cols].corr()

print(correlation_matrix_full)
```
